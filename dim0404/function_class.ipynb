```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
def plot_xy(x,y, title):
    fig, ax = plt.subplots(1, figsize=(10,5))
    ax.plot(x,y)
    ax.set_title(title)
    plt.show()
```

```python
def fun(x):
    if x <= -1:
        return x + 1
    elif x <= 2:
        return -x**2 + 2
    else:
        return x
```

```python
title = 'pag 15'
x = np.linspace(-5, 5, 400)
y = np.array([fun (xi) for xi in x])
```

```python
plot_xy(x, y, title)
```

```python
x1= np.arange(-5, 5)
limits = [limit(fun(xi), x1, -1) for xi in x1]

for i in limits: 
    print(i)
```

```python
x2= np.arange(-5, 5)
limits2 = [limit(fun(xi), x2, 0) for xi in x2]

for i in limits2: 
    print(i)
```
