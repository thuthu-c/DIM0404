```python
import numpy as np
import sympy
import matplotlib.pyplot as plt
```

```python
x = np.array([i for i in range(-100, 101)])
```

```python
y = np.cbrt(x) + x**3 + 7
```

```python
title = 'f(x) = x**(1/3) + x**3+7'
def plot_xy(x,y, title):
    fig, ax = plt.subplots(1, figsize=(10,5))
    ax.plot(x,y)
    ax.set_title(title)
    plt.show()
```

```python
plot_xy(x,y, title)
```

```python
a = np.array([i for i in range(-100, 100)]) 
```

```python
b = np.abs(a)
```

```python
title = 'abs'
plot_xy(a,b, title)
```

```python
n = np.arange(-50, 51, 0.1)
```

```python
m = np.sin(n)
```

```python
title = 'sin(x)' 
plot_xy(n, m, title)
```

```python
t = np.arange(1, 51, 0.1)
```

```python
h = np.log(t)
```

```python
ti = 'ln'
```

```python
plot_xy(t, h, ti)
```

```python
i = np.arange(-1, 1, 0.1)
```

```python
j = np.exp(i)
```

```python
u = 'exp'

plot_xy(i, j, u)
```
